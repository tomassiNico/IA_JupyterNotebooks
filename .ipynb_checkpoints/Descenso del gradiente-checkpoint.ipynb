# Descenso del gradiente 

```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```python
#función anónima con lambda
func = lambda th: np.sin(1 / 2 * th[0] ** 2 - 1 / 4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1])

n = 100

## np.linspace(a, b, n) un vector de n números entre a y b.
X = np.linspace(-2, 2, n) 
Y = np.linspace(-2, 2, n)

## matriz de resultado de función
Z = np.zeros((n, n))


for ix, x in enumerate(X):
    for iy, y in enumerate(Y):
        Z[iy, ix] = func([x, y])

        
plt.contourf(X, Y, Z, 100)
plt.colorbar()

theta = np.random.rand(2) * 4 - 2

t = np.copy(theta)

h = 0.01
lr = 0.001 ##learning rate
gradiente = np.zeros(2)

plt.plot(theta[0], theta[1], "o", color="white")


for _ in range(10000):
    for it, th in enumerate(theta):

        t = np.copy(theta)
        
        t[it] = t[it] + h

        derivada = (func(t) - func(theta) / h)

        gradiente = derivada
        

    theta = theta - gradiente * lr
    
    if(_% 10 == 0):
        plt.plot(theta[0], theta[1], ".", color="red")
    

plt.show()
```
