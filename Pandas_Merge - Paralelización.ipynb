# Merge sobre dataFrame con estructura tabla relacional

```python
import pandas as pd
consumidores = [('A','Panda'),('B','Cocodrilo'),('A','Hormiga')
               ,('F','Sapo'),('C','Cocodrilo'),('A','Gato')]

labels_con = ['Consumidor','Animal']

con_df = pd.DataFrame.from_records(consumidores, columns=labels_con)

animales = [('Panda','China'),('Cocodrilo','Brasil'),('Hormiga','Alemana')
         ,('Perro','Argentino')]

labels_ani = ['Animal','Origen']

ani_df = pd.DataFrame.from_records(animales, columns=labels_ani)
```

```python
pd.merge(con_df,ani_df,on="Animal",how="outer")
```

```python
pd.merge(con_df,ani_df,on="Animal",how="inner")
```

```python
pd.merge(con_df,ani_df,on="Animal",how="right")
```

```python
pd.merge(con_df,ani_df,on="Animal",how="left")
```

```python
from sklearn.datasets import load_iris
import numpy as np
from joblib import Parallel, delayed
```

```python
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head(10)
```

```python
def max_medida(fila):
    if not np.isnan(fila).any():
        names = iris.feature_names
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
results = []
for fila in iris_df.values.tolist():
    results.append(max_medida(fila))
```

```python
results
```

```python
result = Parallel(n_jobs=2, backend="multiprocessing")(
            map(delayed(max_medida), iris_df.values.tolist()))
```
