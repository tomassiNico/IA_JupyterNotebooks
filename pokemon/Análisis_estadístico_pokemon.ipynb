```python
data = pd.read_csv('Pokemon.csv')
data.head(10)
```

## Ahora que ya cargamos los datos analicemos las medidas descriptivas de los pokemons

```python
data.describe()
```

####  Vamos a premiar a algunos pokemos =D

```python
resumen = data.describe()
resumen
```

```python
# el que tiene más aguante 
name = data[data['HP']==resumen.HP['max']]['Name'].values[0]
value = resumen.HP['max']
print('Felicitaciones a {} por ser el pokemon con más HP: {}'.format(name, value))

name=data[data['Attack']==resumen.Attack['max']]['Name'].values[0]
value = resumen.Attack['max']
print('Felicitaciones a {} por ser el pokemon con más Ataque: {}'.format(name, value))

name=data[data['Defense']==resumen.Defense['min']]['Name'].values[0]
value = resumen.Defense['min']
print('Lo sentimos por {} por ser el pokemon con menos defensa: {}'.format(name, value))
```

```python
#cantidad de pokemons por tipo
data['Type 1'].value_counts().plot(kind='bar', figsize=(15,10), title='Cantidad de pokemons categorizados por tipo')
```
