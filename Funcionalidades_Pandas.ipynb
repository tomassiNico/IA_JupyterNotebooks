```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
#cargo los datos en un dataFrame
df = pd.DataFrame(iris.data)
```

```python
#obtengo los 5 primeros registros
df.head()
```

```python
#cargo el nombre de las columnas del dataFrame
df.columns = iris.feature_names
df.head(10)
```

```python
#agrego una columna
df["FLOWER"] = iris.target
df.head()
```

```python
#ultimos registros
df.tail()
```

```python
#desordena el dataFrame, frac es la fraccion de datos que queremos del total.
# frac = 1 significa todos los datos, frac = .5 es la mitad de datos
#toma ejemplos dentro del DataFrame
#util para desordenarlo
df = df.sample(frac= 1)
df.head()
```

```python
#para obtener los datos de una sola columna
df["sepal length (cm)"].head()
```

```python
#tipos de datos de el DataFrame
df.dtypes
```

```python
# si queremos obtener los values para operar con ellos
# es mas facil de trabajar para aplicar operaciones matriciales
df.values
```

```python
#crear otro dataFrame a partir de uno
df2 = df.head(10)
df2
```

```python
#si queremos hacer un filtrado por filas
# ej de fila 10 a fila 20
df[10:20]
```

```python
#podemos aplicar distintas condiciones lógicas para filtrar los datos
df[df["FLOWER"] == 2].head()
```

```python
#otro ejemplo  con condicion OR
df[(df["sepal length (cm)"] > 7) | (df["sepal width (cm)"] == 3.0)].head()
```

```python
#otro filtro
df[pd.isna(df["sepal length (cm)"])]
```

```python
#otro filtro
df[df.FLOWER.isin([1, 2])].head()
```

# Datos duplicados y faltantes

```python
dfduplicada = df.append(df)
dfduplicada.shape
```

```python
dfduplicada = dfduplicada.sample(frac=1)
dfduplicada.shape
```

```python
dfduplicada.head(10)
```

```python
dfclean = dfduplicada.drop_duplicates()
dfclean.shape
```

```python
#para una columna en especial eliminar duplicados
dfclean.drop_duplicates(subset = "FLOWER")
```

```python
#eliminar datos faltantes
#en este caso todos los datos tienen valores así que altera dataFrame
dfclean.dropna().shape

#cuantas columnas queremos con datos
dfclean.dropna(thresh=len(df.columns)-2).head()

# para algunas columnas 
dfclean.dropna(subset = ["FLOWER", "sepal width (cm)"]).head()
```
